Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("kaggle_survey_2021_responses.csv")
df.head()
```

```python
df['Q1'].value_counts()
```

```python
fig,ax1=plt.subplots(1,1,figsize=(24,8))
sns.barplot(x=df['Q1'].value_counts().index,y=df['Q1'].value_counts().values,color="orangered",label="Age_Distribution")
for index, value in enumerate(df['Q1'].value_counts().values):
    ax1.annotate(value,xy=(index,value+100),ha="center", va="center",fontsize=18)
sns.lineplot(x=df['Q1'].value_counts().index,y=df['Q1'].value_counts().values.mean(),color="black",label="Age_Distribution")
plt.legend(fontsize=20)
plt.xlabel("Age_Group", fontsize=20)
plt.ylabel("Age_Count", fontsize=20)
plt.title("Age in 2021",fontsize=30)
plt.show()
```
